```
data = []
choice = 'random'

def line():
    print('---------------------------')

def show_menu():
    line()
    print('MENU')
    line()
    print('1. Add Course')
    print('2. Unenroll Course')
    print('3. View enrolled course list')
    print('4. Exit')
    line()
    choice = input('Enter your choice: ')
    return choice

while choice != '4':
    choice = show_menu()
    if choice == '1':
        item = input('What course you need to enroll ')
        data.append(item)
        print('Course', item, 'added')
    elif choice == '2':
        item = input('Which course you wantto enroll ')
        if item in data:
            data.remove(item)
            print('Course Unenrolled', item)
        else:
            print('Could not find this course', item)
    elif choice == '3':
        print('Your enrolled course list:')
        for item in data:
            print(item)
    elif choice == '4':
        print('Thank you')
    else:
        print('Please enter one of 1, 2, 3 or 4')
```
